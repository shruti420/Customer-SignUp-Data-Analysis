```
import pandas as pd
```

```
df = pd.read_csv('customer_signups.csv')  #importing data
df
```

```
df['signup_date'] = pd.to_datetime(df['signup_date'], errors='coerce')  #Converting signup_date to datetime
df
```

```
# Remove duplicates based on customer_id
df.drop_duplicates(subset='customer_id', inplace=True)
df
```

```
# Standardize values
df['plan_selected'] = df['plan_selected'].str.strip().str.title()
df['gender'] = df['gender'].str.strip().str.title()
df['source'] = df['source'].str.strip().str.title()
df
```

```
# Count missing values
missing_values = df.isnull().sum()
missing_percent = (df.isnull().mean() * 100).round(2)

missing_values
missing_percent

missing_summary = pd.DataFrame({
    'Missing Count': df.isnull().sum(),
    'Missing %': round((df.isnull().sum() / len(df)) * 100, 2)
})
print(missing_summary)
```

```
# For text columns like 'region' or 'gender':
df['region'].fillna('Unknown', inplace=True)
df['gender'].fillna('Not Specified', inplace=True)

df['signup_date'] = pd.to_datetime(df['signup_date'], errors='coerce')
df

df['age'] = df['age'].fillna(df['age'].median())
df=df.replace('NaN', '')
df=df.replace('??', '')
df.fillna(' ') 
```

```
# Weekly sign-ups
weekly_signups = df.resample('W-Mon', on='signup_date').size()
weekly_signups
```

```
# Sign-ups by source, region, plan
source_counts = df['source'].value_counts()
region_counts = df['region'].value_counts()
plan_counts = df['plan_selected'].value_counts()
source_counts
region_counts
plan_counts
```

```
# Marketing opt-in by gender
optin_by_gender = df.groupby('gender')['marketing_opt_in'].value_counts()
optin_by_gender
```

```
# Age summary
age_stats = df['age'].describe()
age_nulls = df['age'].isnull().sum()
age_stats
age_nulls
```

```

import matplotlib.pyplot as plt


# Sign-ups per week
signups_per_week = df.set_index('signup_date').resample('W')['customer_id'].count()

# Plot
plt.figure(figsize=(10, 5))
signups_per_week.plot(kind='bar', color='skyblue')
plt.title("Sign-Ups Per Week")
plt.xlabel("Week")
plt.ylabel("Number of Sign-Ups")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```
# Convert age to numeric
df['age'] = pd.to_numeric(df['age'], errors='coerce')

# Categorize age groups
df['age_group'] = pd.cut(df['age'], bins=[0, 24, 35, 50, 100], labels=['<25', '25-35', '36-50', '51+'])

# Standardize plan names
df['plan_selected'] = df['plan_selected'].str.strip().str.title()

# Group and plot
plan_age = df.groupby(['age_group', 'plan_selected']).size().unstack().fillna(0)

# Plot
plan_age.plot(kind='bar', stacked=True, figsize=(8, 5))
plt.title("Plan Selection by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Sign-Ups")
plt.tight_layout()
plt.show()
```

```
# Clean gender and marketing opt-in columns
df['gender'] = df['gender'].str.strip().str.title().fillna('Not Specified')
df['marketing_opt_in'] = df['marketing_opt_in'].str.strip().str.title().fillna('No')

# Group and plot
optin_gender = df.groupby(['gender', 'marketing_opt_in']).size().unstack().fillna(0)

optin_gender.plot(kind='bar', figsize=(8, 5))
plt.title("Marketing Opt-In by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Users")
plt.tight_layout()
plt.show()
```
